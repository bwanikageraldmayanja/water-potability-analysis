```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading csv
df = pd.read_csv('/content/drive/MyDrive/ML with Gero/water_potability.csv')
df.head(30)
```

```python
df.info()
```

```python
df.describe()
```

```python
#identifying missing data
df.isnull().sum()
```

We realise that ph, sulfate and Trihalomethanes are the columns with the missing data

```python
#filling the missing values in the ph and sulfate columns with the mean of their respective columns.
ph_new = df['ph'].fillna(df['ph'].mean(), inplace=True)
sulfate_new = df['Sulfate'].fillna(df['Sulfate'].mean(), inplace=True)
```

```python
#confirming whether the mssing values have been replaced with the mean in the required columns
df.head(30)
```

```python
#checking for duplicates
df.duplicated().sum()
```

This confirms that there are no duplicates

```python
#Creating a new column called ph_level Classify the pH values into categories: "Acidic" (pH < 6.5), "Neutral" (6.5 <= pH <= 7.5), and "Alkaline" (pH > 7.5)
df['ph_level'] = pd.cut(df['ph'], bins=[0, 6.5, 7.5, np.inf], labels=['Acidic', 'Neutral', 'Alkaline'])
df.head()
```

```python
#groupby() to compare potable and non-potable water.
df.groupby('Potability')[['Hardness', 'Chloramines']].mean()
```

```python
#row of the water sample with the highest solids concentration
df[df['Solids'] == df['Solids'].max()]
```

```python
#the ph and Hardness of the water sample with the highest Solids concentration
df[df['Solids'] == df['Solids'].max()][['ph', 'Hardness']]
```

```python
#saving the cleaned DataFrame to a new file named cleaned_water_potability.csv
df.to_csv('cleaned_water_potability.csv', index=False)
```

#Part 2: Data Visualization and Storytelling


```python
# Creating 2x2 dashboard
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
# A bar chart (or countplot) showing the count of "Potable" (1) vs. "Not Potable" (0) water samples
sns.countplot(x='Potability', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Potability Distribution')
# A histogram to show the distribution of pH values across all samples.
sns.histplot(df['ph'], ax=axes[0, 1])
axes[0, 1].set_title('pH Levels Distribution')
# A box plot to compare the distribution of Sulfate for potable and non-potable water.
sns.boxplot(x='Potability', y='Sulfate', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Sulfate Content by Potability')
# A scatter plot to investigate the relationship between Hardness and Organic_carbon.
sns.scatterplot(x= 'Hardness', y='Organic_carbon', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Hardness vs. Organic Carbon')
#Adding a main title to the entire 2x2 dashboard.
fig.suptitle('Water Potability Analysis Report', fontsize=16)
```

Findings from the 'Potability Distribution' we see that our dataset is imbalanced and from the 'pH level Distribution' shows a strong concetration around 6-8 with a peak near 7.

We see that the mean of both potable and non potable is almost the same. which shows that sulfate alone does not show the distinction between the two hence may not be a clear factor that affects potability. meanwhile, the 'organic_carbon Vs Hardness' shows a dense clustering of the data points.

In summary, The Potability Distribution shows an imbalance favoring potable conditions. The ‘pH Levels Distribution’ peaks at 6-8, highlighting neutral pH as a critical factor. The ‘Sulphate Content by Potability’ plot shows overlapping levels, suggesting sulfate is not a strong indicator/differentiator. The ‘Hardness Vs Organic Carbon’ scatter plot’s dense cluster hints at a potential relationship influencing potability. Ultimately, The data story shows that pH is a primary driver, with Hardness and Organic Carbon interplay needing further exploration, considering the imbalance.
